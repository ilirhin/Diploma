```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib

font = {'family' : 'normal',
        'size'   : 15}

matplotlib.rc('font', **font)
```

```python
def mplot(suffix, title, mod):

    with open('opt_reg_LR' + suffix + '.txt', 'r') as fi:
        opt_LR_values = np.array(map(float, fi.read().strip().split()))

    with open('opt_reg_R' + suffix + '.txt', 'r') as fi:
        opt_R_values = np.array(map(float, fi.read().strip().split()))

    with open('upg_reg_LR' + suffix + '.txt', 'r') as fi:
        upg_LR_values = np.array(map(float, fi.read().strip().split()))

    with open('upg_reg_R' + suffix + '.txt', 'r') as fi:
        upg_R_values = np.array(map(float, fi.read().strip().split()))

    with open('simple_reg_LR' + suffix + '.txt', 'r') as fi:
        simple_LR_values = np.array(map(float, fi.read().strip().split()))

    with open('simple_reg_R' + suffix + '.txt', 'r') as fi:
        simple_R_values = np.array(map(float, fi.read().strip().split()))


    indices = range(len(opt_LR_values))
    plt.figure(figsize=(10, 6))
    plt.title('L + R')
    plt.ylabel('L + R')
    plt.xlabel('iteration')
    plt.plot(indices[1:], opt_LR_values[1:], linewidth=2.50)
    plt.plot(indices[1:], upg_LR_values[1:], linewidth=4.50)
    plt.plot(indices[1:], simple_LR_values[1:], linewidth=2.50)
    plt.legend(['gradient', 'unbiased', 'original'], loc='lower right')
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.title('L + R')
    plt.plot(indices[30:], opt_LR_values[30:], linewidth=2.50)
    plt.plot(indices[30:], upg_LR_values[30:], linewidth=4.50)
    plt.plot(indices[30:], simple_LR_values[30:], linewidth=2.50)
    plt.ylabel('L + R')
    plt.xlabel('iteration')
    plt.legend(['gradient', 'unbiased', 'original'], loc='lower right')
    plt.show()
    

    plt.figure(figsize=(10, 6))
    plt.title('R')
    plt.plot(indices[1:], opt_R_values[1:] * mod, linewidth=2.50)
    plt.plot(indices[1:], upg_R_values[1:] * mod, linewidth=2.50)
    plt.plot(indices[1:], simple_R_values[1:] * mod, linewidth=2.50)
    plt.ylabel('R')
    plt.xlabel('iteration')
    plt.legend(['gradient', 'unbiased', 'original'], loc='upper right')
    plt.show()
```

```python
mplot('', '', 1e6)
```

```python
mplot('_5', '', 1e5)
```

```python
mplot('_8', '', 1e8)
```
